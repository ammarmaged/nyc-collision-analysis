```python
# calculate percentage of missing values in crash dataset
print("Missing values percentage in df_crashes:")
print((df_crashes.isnull().sum() / len(df_crashes) * 100).sort_values(ascending=False))
```

```python
# Create a new binary feature 'IS_3_VEHICLE_CRASH' based on the presence of a third contributing factor or vehicle type
df_crashes['IS_3_VEHICLE_CRASH'] = ((df_crashes['CONTRIBUTING FACTOR VEHICLE 3'].notnull()) | (df_crashes['VEHICLE TYPE CODE 3'].notnull())).astype(int)

print("New combined binary feature 'IS_3_VEHICLE_CRASH' added to df_crashes.")
print(df_crashes[['CONTRIBUTING FACTOR VEHICLE 3', 'VEHICLE TYPE CODE 3', 'IS_3_VEHICLE_CRASH']].head())
```

```python
# Identify columns to drop from df_crashes based on >90% missing
missing_percentages_crashes = df_crashes.isnull().sum() / len(df_crashes) * 100
columns_to_drop_crashes = missing_percentages_crashes[missing_percentages_crashes > 90].index.tolist()

# Explicitly add 'ZIP CODE' to the list of columns to drop if not already present
if 'ZIP CODE' not in columns_to_drop_crashes:
    columns_to_drop_crashes.append('ZIP CODE')

print(f"Columns to drop from df_crashes (more than 90% missing or explicitly requested): {columns_to_drop_crashes}")

# Drop columns from df_crashes
df_crashes = df_crashes.drop(columns=columns_to_drop_crashes)

print(f"df_crashes shape after dropping columns: {df_crashes.shape}")
```

```python
# Impute missing values in 'VEHICLE TYPE CODE 2' with 'No Second Vehicle'
df_crashes['VEHICLE TYPE CODE 2'] = df_crashes['VEHICLE TYPE CODE 2'].fillna('No Second Vehicle')

# Impute missing values in 'CONTRIBUTING FACTOR VEHICLE 2' with 'Unspecified'
df_crashes['CONTRIBUTING FACTOR VEHICLE 2'] = df_crashes['CONTRIBUTING FACTOR VEHICLE 2'].fillna('Unspecified')

print("Missing values in 'VEHICLE TYPE CODE 2' and 'CONTRIBUTING FACTOR VEHICLE 2' have been imputed.")
print(df_crashes[['VEHICLE TYPE CODE 2', 'CONTRIBUTING FACTOR VEHICLE 2']].head())
```

```python
# Calculate current missing percentages for df_crashes
missing_percentages_current = df_crashes.isnull().sum() / len(df_crashes) * 100

# Identify columns with missing values > 0% and <= 1%
columns_to_drop_rows = missing_percentages_current[(missing_percentages_current > 0) & (missing_percentages_current <= 1)].index.tolist()

if columns_to_drop_rows:
    print(f"Columns with missing values between 0% and 1% : {columns_to_drop_rows}")
    print(f"df_crashes shape before dropping rows: {df_crashes.shape}")

    # Drop rows where any of these identified columns have missing values
    df_crashes.dropna(subset=columns_to_drop_rows, inplace=True)

    print(f"df_crashes shape after dropping rows for columns with <1% missing values: {df_crashes.shape}")
    print("Missing values in affected columns after dropping rows:")
    print(df_crashes[columns_to_drop_rows].isnull().sum())
else:
    print("No columns found with missing values between 0% and 1% (exclusive of 0, inclusive of 1%).")
```

```python
import numpy as np

# Combine street name columns into a single 'STREET NAME' column with prioritization.
# This prioritizes 'ON STREET NAME'. If it's null, it tries 'CROSS STREET NAME'.
# If 'CROSS STREET NAME' is also null, it tries 'OFF STREET NAME'.
# If all three are null, 'STREET NAME' will be NaN.
df_crashes['STREET NAME'] = df_crashes['ON STREET NAME'].fillna(df_crashes['CROSS STREET NAME']).fillna(df_crashes['OFF STREET NAME'])

print("New 'STREET NAME' column created by combining street names with prioritization.")
print(df_crashes[['ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME', 'STREET NAME']].head())

# Drop the original street name columns
df_crashes.drop(columns=['ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME'], inplace=True)
print("Original street name columns dropped.")

# Impute missing values in 'STREET NAME' with 'UNKNOWN'
df_crashes['STREET NAME'].fillna('UNKNOWN', inplace=True)
print("Missing values in 'STREET NAME' imputed with 'UNKNOWN'.")
print(df_crashes[['STREET NAME']].head())
```

```python
def fill_geographic_data(df):
    """Fill missing BOROUGH using LOCATION and STREET NAME with mode fallback."""
    df = df.copy()

    # --- 1. Create LOCATION → BOROUGH mapping ---
    location_to_borough = (
        df[df['BOROUGH'].notna() & df['LOCATION'].notna()]
        .groupby('LOCATION')['BOROUGH']
        .agg(lambda x: x.mode()[0] if not x.mode().empty else None)
        .to_dict()
    )

    # --- 2. Create STREET NAME → BOROUGH mapping ---
    street_to_borough = (
        df[df['BOROUGH'].notna() & df['STREET NAME'].notna()]
        .groupby('STREET NAME')['BOROUGH']
        .agg(lambda x: x.mode()[0] if not x.mode().empty else None)
        .to_dict()
    )

    # --- 3. Fill missing BOROUGH using LOCATION ---
    mask_location = df['BOROUGH'].isna() & df['LOCATION'].notna()
    df.loc[mask_location, 'BOROUGH'] = df.loc[mask_location, 'LOCATION'].map(location_to_borough)

    # --- 4. Fill remaining missing BOROUGH using STREET NAME ---
    mask_street = df['BOROUGH'].isna() & df['STREET NAME'].notna()
    df.loc[mask_street, 'BOROUGH'] = df.loc[mask_street, 'STREET NAME'].map(street_to_borough)

    # --- 5. Fill any remaining missing values with 'UNKNOWN' ---
    df['BOROUGH'] = df['BOROUGH'].fillna('UNKNOWN')

    return df

# Apply the improved function
df_crashes = fill_geographic_data(df_crashes)

# Check results
print("Missing values after imputation:")
print(f"BOROUGH: {df_crashes['BOROUGH'].isnull().sum()} ({df_crashes['BOROUGH'].isnull().mean()*100:.2f}%)")
print(f"\nBOROUGH value counts:\n{df_crashes['BOROUGH'].value_counts()}")
```

```python
import pandas as pd

# Replace 'UNKNOWN' with 'QUEENS' in the 'BOROUGH' column as Queens is the mode

df_crashes['BOROUGH'] = df_crashes['BOROUGH'].replace('UNKNOWN', 'QUEENS')

print(" 'UNKNOWN' values in 'BOROUGH' column replaced with 'QUEENS'.")
print(df_crashes['BOROUGH'].value_counts(dropna=False))
```

```python
# Impute missing LATITUDE and LONGITUDE using the median of their respective BOROUGH

# Calculate median LATITUDE and LONGITUDE for each BOROUGH
median_coords_by_borough = df_crashes.groupby('BOROUGH')[['LATITUDE', 'LONGITUDE']].median()

# Fill missing LATITUDE
df_crashes['LATITUDE'] = df_crashes.apply(
    lambda row: median_coords_by_borough.loc[row['BOROUGH'], 'LATITUDE']
    if pd.isna(row['LATITUDE']) and row['BOROUGH'] in median_coords_by_borough.index
    else row['LATITUDE'],
    axis=1
)

# Fill missing LONGITUDE
df_crashes['LONGITUDE'] = df_crashes.apply(
    lambda row: median_coords_by_borough.loc[row['BOROUGH'], 'LONGITUDE']
    if pd.isna(row['LONGITUDE']) and row['BOROUGH'] in median_coords_by_borough.index
    else row['LONGITUDE'],
    axis=1
)

# For any remaining missing LATITUDE/LONGITUDE (e.g., if a borough itself had all missing values or was 'UNKNOWN'),
# fill with the global median to ensure no NaNs remain for these columns.
# This should only happen for 'UNKNOWN' BOROUGH which might not have a median computed, or if a whole borough has NaNs.
# From previous steps, we already filled 'UNKNOWN' BOROUGHs, so this step might not be strictly necessary, but good for robustness.
median_latitude_global = df_crashes['LATITUDE'].median()
median_longitude_global = df_crashes['LONGITUDE'].median()
df_crashes['LATITUDE'].fillna(median_latitude_global, inplace=True)
df_crashes['LONGITUDE'].fillna(median_longitude_global, inplace=True)

print("Missing values in 'LATITUDE' and 'LONGITUDE' imputed based on Borough medians.")
print("Null counts after imputation:")
print(df_crashes[['LATITUDE', 'LONGITUDE']].isnull().sum())
```

```python
import numpy as np

print(f"'STREET NAME' Unknown values before imputation: {df_crashes[df_crashes['STREET NAME'] == 'UNKNOWN'].shape[0]}")

# Temporarily replace 'UNKNOWN' with NaN to use fillna/transform effectively
df_crashes['STREET NAME'] = df_crashes['STREET NAME'].replace('UNKNOWN', np.nan)

# Use groupby and transform to fill NaNs with the mode of 'STREET NAME' within each geographic group
# This ensures we are filling with the most common street name for that specific coordinate and borough.
# The lambda function handles cases where a group might only contain NaNs or be empty
df_crashes['STREET NAME'] = df_crashes.groupby(['BOROUGH', 'LATITUDE', 'LONGITUDE'])['STREET NAME'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)

# Convert any remaining NaNs back to 'UNKNOWN' (these are groups where no known street name was available)
df_crashes['STREET NAME'] = df_crashes['STREET NAME'].fillna('UNKNOWN')

print(f"'STREET NAME' Unknown values after coordinate-based imputation: {df_crashes[df_crashes['STREET NAME'] == 'UNKNOWN'].shape[0]}")

# Fallback to Borough mode for any remaining 'UNKNOWN' (less granular, but better than nothing)
if df_crashes[df_crashes['STREET NAME'] == 'UNKNOWN'].shape[0] > 0:
    # Calculate mode street name per borough, ignoring 'UNKNOWN' values
    borough_street_mode = df_crashes[df_crashes['STREET NAME'] != 'UNKNOWN'].groupby('BOROUGH')['STREET NAME'].agg(
        lambda x: x.mode()[0] if not x.mode().empty else np.nan
    )
    # Apply this mode imputation only to rows where 'STREET NAME' is still 'UNKNOWN'
    # Temporarily replace 'UNKNOWN' with NaN for fillna
    df_crashes['STREET NAME'] = df_crashes['STREET NAME'].replace('UNKNOWN', np.nan)
    df_crashes['STREET NAME'] = df_crashes.groupby('BOROUGH')['STREET NAME'].transform(
        lambda x: x.fillna(borough_street_mode.get(x.name, np.nan)) # x.name is the borough value
    )
    df_crashes['STREET NAME'] = df_crashes['STREET NAME'].fillna('UNKNOWN') # Convert remaining NaNs back

    print(f"'STREET NAME' Unknown values after borough-based fallback imputation: {df_crashes[df_crashes['STREET NAME'] == 'UNKNOWN'].shape[0]}")

print("Imputation for 'STREET NAME' using geographical coordinates and borough mode complete.")
print("Value counts for 'STREET NAME' after imputation:")
print(df_crashes['STREET NAME'].value_counts().head(10))
```

Detect outliers in longitude and latitiude

```python



import pandas as pd

def calculate_outliers_iqr(series):
    """
    Calculates outliers for a given pandas Series using the IQR method.

    Args:
        series (pd.Series): The input Series for which to calculate outliers.

    Returns:
        pd.Series: A boolean Series indicating which values are outliers.
    """
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return (series < lower_bound) | (series > upper_bound)

print("Function `calculate_outliers_iqr` defined.")

unique_boroughs = df_integrated['BOROUGH'].unique()
borough_outliers = {}

for borough in unique_boroughs:
    # Filter data for the current borough
    borough_df = df_integrated[df_integrated['BOROUGH'] == borough].copy()

    # Calculate latitude outliers
    latitude_outliers = calculate_outliers_iqr(borough_df['LATITUDE'])

    # Calculate longitude outliers
    longitude_outliers = calculate_outliers_iqr(borough_df['LONGITUDE'])

    # Store the boolean series in the dictionary
    borough_outliers[borough] = {
        'latitude_outliers': latitude_outliers,
        'longitude_outliers': longitude_outliers
    }

print("Outlier calculation by borough complete. Results stored in `borough_outliers` dictionary.")

outlier_percentages = {}

for borough, outliers_data in borough_outliers.items():
    total_records = len(df_integrated[df_integrated['BOROUGH'] == borough])

    # Calculate percentage of latitude outliers
    lat_outlier_count = outliers_data['latitude_outliers'].sum()
    lat_outlier_percentage = (lat_outlier_count / total_records) * 100 if total_records > 0 else 0

    # Calculate percentage of longitude outliers
    lon_outlier_count = outliers_data['longitude_outliers'].sum()
    lon_outlier_percentage = (lon_outlier_count / total_records) * 100 if total_records > 0 else 0

    outlier_percentages[borough] = {
        'total_records': total_records,
        'latitude_outliers_count': lat_outlier_count,
        'latitude_outlier_percentage': lat_outlier_percentage,
        'longitude_outliers_count': lon_outlier_count,
        'longitude_outlier_percentage': lon_outlier_percentage
    }
```

```python
outlier_percentages = {}

total_records_integrated = len(df_integrated)

for borough, outliers_data in borough_outliers.items():
    # The total records for the current borough (for context, as per instruction 7)
    borough_total_records = len(df_integrated[df_integrated['BOROUGH'] == borough])

    # Calculate count of latitude outliers for this borough
    lat_outlier_count = outliers_data['latitude_outliers'].sum()
    # Calculate percentage using total_records_integrated as denominator
    lat_outlier_percentage = (lat_outlier_count / total_records_integrated) * 100 if total_records_integrated > 0 else 0

    # Calculate count of longitude outliers for this borough
    lon_outlier_count = outliers_data['longitude_outliers'].sum()
    # Calculate percentage using total_records_integrated as denominator
    lon_outlier_percentage = (lon_outlier_count / total_records_integrated) * 100 if total_records_integrated > 0 else 0

    outlier_percentages[borough] = {
        'total_records_in_borough': borough_total_records, # Storing borough-specific total for context
        'latitude_outliers_count': lat_outlier_count,
        'latitude_outlier_percentage': lat_outlier_percentage,
        'longitude_outliers_count': lon_outlier_count,
        'longitude_outlier_percentage': lon_outlier_percentage
    }

print("Geographical Outlier Percentages by Borough (relative to total dataset):")
for borough, data in outlier_percentages.items():
    print(f"\nBorough: {borough}")
    print(f"  Total Records (in borough): {data['total_records_in_borough']}")
    print(f"  Latitude Outliers (count): {data['latitude_outliers_count']}")
    print(f"  Latitude Outliers (% of total dataset): {data['latitude_outlier_percentage']:.2f}% ")
    print(f"  Longitude Outliers (count): {data['longitude_outliers_count']}")
    print(f"  Longitude Outliers (% of total dataset): {data['longitude_outlier_percentage']:.2f}% ")

# Calculate and print the sum of these percentages
total_latitude_outlier_percentage = sum(data['latitude_outlier_percentage'] for data in outlier_percentages.values())
total_longitude_outlier_percentage = sum(data['longitude_outlier_percentage'] for data in outlier_percentages.values())

print(f"\nTotal Latitude Outliers (sum of borough percentages): {total_latitude_outlier_percentage:.2f}% ")
print(f"Total Longitude Outliers (sum of borough percentages): {total_longitude_outlier_percentage:.2f}% ")
```

```python
import pandas as pd
import numpy as np

# 1. Store the initial shape of the df_integrated DataFrame
initial_shape = df_integrated.shape
print(f"Initial shape of df_integrated: {initial_shape}")

# 2. Initialize a pandas Series named combined_outlier_mask
combined_outlier_mask = pd.Series(False, index=df_integrated.index)

# 3. Loop through each borough and its corresponding outlier data
for borough, outliers_data in borough_outliers.items():
    # a. Get the indices of df_integrated that correspond to the current borough.
    borough_indices = df_integrated[df_integrated['BOROUGH'] == borough].index

    # b. Extract the latitude_outliers and longitude_outliers Series
    lat_outliers_series = outliers_data['latitude_outliers']
    lon_outliers_series = outliers_data['longitude_outliers']

    # c. Create full-sized boolean Series for latitude and longitude outliers
    full_lat_mask_for_borough = pd.Series(False, index=df_integrated.index)
    full_lat_mask_for_borough.loc[borough_indices] = lat_outliers_series

    full_lon_mask_for_borough = pd.Series(False, index=df_integrated.index)
    full_lon_mask_for_borough.loc[borough_indices] = lon_outliers_series

    # d. Combine the full_lat_mask_for_borough and full_lon_mask_for_borough
    borough_combined_outlier_mask = full_lat_mask_for_borough | full_lon_mask_for_borough

    # e. Update the overall combined_outlier_mask
    combined_outlier_mask = combined_outlier_mask | borough_combined_outlier_mask

# 5. Create a new DataFrame df_integrated_filtered by removing outlier rows
df_integrated_filtered = df_integrated[~combined_outlier_mask].copy()

# 6. Store the shape of df_integrated_filtered
final_shape = df_integrated_filtered.shape

# 7. Print the initial and final shapes, and the number of rows removed.
print(f"Final shape of df_integrated after filtering outliers: {final_shape}")
removed_rows = initial_shape[0] - final_shape[0]
print(f"Number of rows removed: {removed_rows}")

# 8. Overwrite the df_integrated DataFrame with df_integrated_filtered.
df_integrated = df_integrated_filtered

# 9. Display the first 5 rows of the updated df_integrated.
print("\nFirst 5 rows of df_integrated after filtering:")
display(df_integrated.head())
```
